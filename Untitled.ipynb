```python
cols = ['H', 'diameter', 'albedo','epoch', 'e', 'a', 'q', 'i','w','tp','per', 'moid_ld']
pp = sns.pairplot(df[cols], height=1.8, aspect=1.8,
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Dependence', fontsize=14)
```

```python
f, ax = plt.subplots(figsize=(20, 14))
corr = df.corr(method='pearson')
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=0.025)
f.subplots_adjust(top=0.93)
t= f.suptitle('Pearson Correlation Heatmap', fontsize=14)
```

```python
f.savefig('Pearson.png')
```

```python
f, ax = plt.subplots(figsize=(20, 14))
corr = df.corr(method='spearman')
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=0.025)
f.subplots_adjust(top=0.93)
t= f.suptitle('Spearman Correlation Heatmap', fontsize=14)
```

```python
f.savefig('Spearman.png')
```

```python
# linear regression model
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

X = pd.DataFrame(df['diameter'])
y = pd.DataFrame(df['albedo'])
model = LinearRegression()
scores = []

kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
  model.fit(X.iloc[train,:], y.iloc[train,:])
  score = model.score(X.iloc[test,:], y.iloc[test,:])
  scores.append(score)
print(scores)
```
